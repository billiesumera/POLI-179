# Method 1 - Word2vec


Using Word2Vec
Model Setup:

Use  pre-trained Word2Vec model from library Gensim, which has been trained on various large datasets (like Google News, Wikipedia, etc.).

Calculate Similarity:

With Word2Vec, you can directly calculate the cosine similarity between vectors of two words.

# Load Libraries and Pre-Trained Model

```python
import matplotlib.pyplot as plt
from scipy.spatial import distance
from sklearn.decomposition import PCA
```

```python
import gensim.downloader as api
from scipy.spatial import distance

# Load pre-trained model
model = api.load('word2vec-google-news-300')
```

```python
# mount drive
from google.colab import drive
drive.mount('/content/drive/', force_remount=True)

from google.colab import files
```

# Profession 1 - Doctor

```python
# Words to compare - Profession 1 - doctor/female
word1 = 'doctor'
females = ['female', 'woman', 'lady', 'girl', 'she', 'her']

# Calculate similarity scores
similarities = []
for female in females:
    if word1 in model.key_to_index and female in model.key_to_index:  # Check if words are in the model
        similarity = 1 - distance.cosine(model[word1], model[female])
        similarities.append(similarity)
        print(f"Similarity between '{word1}' and '{female}': {similarity:.3f}")
    else:
        similarities.append(0)  # Handle cases where the word might not be in the model
```

```python
# Words to compare - Profession 1 - doctor/male
word1 = 'doctor'
males = ['male', 'man', 'guy', 'boy', 'he', 'him']
# Calculate similarity scores
similarities = []
for male in males:
    if word1 in model.key_to_index and male in model.key_to_index:  # Check if words are in the model
        similarity = 1 - distance.cosine(model[word1], model[male])
        similarities.append(similarity)
        print(f"Similarity between '{word1}' and '{male}': {similarity:.3f}")
    else:
        similarities.append(0)  # Handle cases where the word might not be in the model
```

```python
words = [word1] + males + females
word_vectors = [model[word] for word in words if word in model.key_to_index]

# PCA for dimensionality reduction to 2D
pca = PCA(n_components=2)
word_vecs_2d = pca.fit_transform(word_vectors)

# Plotting
plt.figure(figsize=(10, 6))
for word, coord in zip(words, word_vecs_2d):
    x, y = coord
    plt.scatter(x, y, marker='x', color='blue')
    plt.text(x + 0.1, y + 0.1, word, fontsize=9)
plt.title('Doctor: 2D Visualization of Word Vectors using PCA')
plt.savefig('PCA Visualization (Doctor).png')
files.download('PCA Visualization (Doctor).png')
plt.show()
```

```python
import nltk
from nltk.corpus import wordnet as wn

nltk.download('wordnet')
```

```python
# Get synsets for 'doctor'
doctor_synsets = wn.synsets('doctor')

# Words to compare
female2 = ['female', 'woman', 'lady', 'girl']

# Find similarities
for word in female2:
    word_synsets = wn.synsets(word)
    for doctor_syn in doctor_synsets:
        for word_syn in word_synsets:
            # Calculate path similarity
            similarity = doctor_syn.path_similarity(word_syn)
            if similarity is not None:
                print(f"Path similarity between {doctor_syn.name()} and {word_syn.name()} is {similarity:.3f}")
```
